## Students scoring above/below average

```python
# Calculate *avg* score for each student.
math_avg = df_1['math score'].mean()
reading_avg = df_1['reading score'].mean()
writing_avg = df_1['writing score'].mean()

print(f'Math Avg: {math_avg:.2f}, Reading Avg: {reading_avg:.2f},  Writing Avg: {writing_avg:.2f}')
```

```python
# Add new column to label scores like `below or above avg`.
df_1['math_lavel'] = df_1['math score'].apply(lambda x: 'Above Avg' if x>math_avg else 'Below Avg')
df_1['reading_lavel'] = df_1['reading score'].apply(lambda x: 'Above Avg' if x>reading_avg else 'Below Avg')
df_1['writing_lavel'] = df_1['writing score'].apply(lambda x: 'Above Avg' if x>writing_avg else 'Below Avg')
```

```python
# Calculate total and average score across all subjects
df_1['total_score'] = df_1[['math score', 'reading score', 'writing score']].sum(axis=1)
df_1['average_score'] = df_1['total_score'] / 3

# Now calculate overall dataset average
overall_avg = df_1['average_score'].mean()

# Label students as above/below overall average
df_1['overall_level'] = df_1['average_score'].apply(lambda x: 'Above Avg' if x > overall_avg else 'Below Avg')
```

## Top 10 students per subject

```python
# Top 10 Students in Maths
top_math = df_1.sort_values(by='math score',ascending=False).head(10)
# Top 10 Students in Reading
top_reading = df_1.sort_values(by='reading score',ascending=False).head(10)
# Top 10 Students in Writing
top_writing = df_1.sort_values(by='writing score',ascending=False).head(10)

# Display names with score
print('Top 10 Math Scorers:/n', top_math[['math score']])
print('Top 10 Reading Scorers:/n', top_reading[['reading score']])
print('Top 10 Writing Scorers:/n', top_writing[['writing score']])
```
